```python
spark.sql("""
SELECT *,
left(customer_name, charindex(' ', customer_name) -1) as f_name,
right(customer_name, charindex(' ', reverse(customer_name)) -1) as l_name,
quantity * unit_price as amount,
current_timestamp as processed_at
FROM 
test.bronze.bronze_table
""").createOrReplaceTempView("silver_source")
```

```python
%sql
CREATE TABLE IF NOT EXISTS test.silver.silver_table
AS
select * from silver_source
```

```python
%sql
MERGE INTO test.silver.silver_table as trg
USING silver_source as src
ON trg.order_id = src.order_id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```

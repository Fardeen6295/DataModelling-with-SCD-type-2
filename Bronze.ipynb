```python
if spark.catalog.tableExists('test.bronze.bronze_table'):
    last_load_date = spark.sql("""SELECT max(order_date) FROM test.bronze.bronze_table""").collect()[0][0]
else:
    last_load_date = '1900-01-01'
```

```python
last_load_date
```

```python
spark.sql(f"""
          SELECT * FROM test.default.source_data WHERE order_date > '{last_load_date}'
          """).createOrReplaceTempView('bronze_source')
```

```python
%sql
CREATE OR REPLACE TABLE test.bronze.bronze_table
AS
SELECT * FROM bronze_source
```
